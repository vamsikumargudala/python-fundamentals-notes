### Shallow Copy vs. Deep Copy (for Nested Lists)

When dealing with nested lists (lists containing other lists), the distinction between shallow and deep copies becomes crucial.

**1. Shallow Copy:**
A shallow copy creates a new list, but it does *not* create new copies of the nested objects within the original list. Instead, it copies references to those nested objects. This means that if you modify a nested object in the copied list, the original list's nested object will also be affected because they both refer to the same underlying object.

Methods like `list.copy()`, `list()`, `[:]` slicing, and `[*list]` (unpacking) all perform shallow copies.

**Example:**
```python
a = [[1, 2], [3, 4]]
b = a.copy() # Shallow copy

b[0][0] = 99
print(f"Original list 'a': {a}") # Output: [[99, 2], [3, 4]] - 'a' was affected!
print(f"Shallow copied list 'b': {b}") # Output: [[99, 2], [3, 4]]
```

**2. Deep Copy:**
A deep copy creates a new list and recursively creates new independent copies of *all* objects found in the original list, including nested objects. This ensures that the copied list and its contents are entirely independent of the original list. Changes to the deep-copied list or its nested elements will not affect the original list.

To perform a deep copy, you typically use the `copy.deepcopy()` function from Python's `copy` module.

**Example:**
```python
import copy

a = [[1, 2], [3, 4]]
c = copy.deepcopy(a) # Deep copy

c[0][0] = 100
print(f"Original list 'a': {a}") # Output: [[1, 2], [3, 4]] - 'a' remains unchanged!
print(f"Deep copied list 'c': {c}") # Output: [[100, 2], [3, 4]]
```

**Summary:**
*   Use **shallow copy** when your list contains only immutable (e.g., numbers, strings, tuples) or non-nested mutable objects, or when you specifically want changes in nested objects to reflect across both lists.
*   Use **deep copy** when your list contains nested mutable objects (e.g., lists, dictionaries) and you need complete independence between the original and the copied structure.

````python
import copy

# Original nested list
a = [[1, 2], [3, 4]]
print(f"Original list 'a': {a}\n")

# --- Shallow Copy ---
print("--- Demonstrating Shallow Copy ---")
b = a.copy() # Using .copy() for shallow copy
print(f"Shallow copied list 'b' (initially): {b}")

b[0][0] = 99 # Modify a nested element in the shallow copy
print(f"Modified 'b[0][0]' to 99.")
print(f"Original list 'a' after modifying 'b': {a}") # 'a' is affected!
print(f"Shallow copied list 'b' after modification: {b}\n")

# --- Deep Copy ---
print("--- Demonstrating Deep Copy ---")
c = copy.deepcopy(a) # Using copy.deepcopy() for deep copy
print(f"Deep copied list 'c' (initially): {c}")

c[0][0] = 100 # Modify a nested element in the deep copy
print(f"Modified 'c[0][0]' to 100.")
print(f"Original list 'a' after modifying 'c': {a}") # 'a' is NOT affected!
print(f"Deep copied list 'c' after modification: {c}\n")
````

````python
# --- Demonstrating Unpacking ([*list]) for Shallow Copy ---
print("--- Demonstrating Unpacking ([*list]) for Shallow Copy ---")

a_original = [[10, 20], [30, 40]]
print(f"Original list 'a_original': {a_original}")

d = [*a_original] # Using unpacking for shallow copy
print(f"Shallow copied list 'd' (initially): {d}")

d[0][0] = 55 # Modify a nested element in the unpacked copy
print(f"Modified 'd[0][0]' to 55.")
print(f"Original list 'a_original' after modifying 'd': {a_original}") # 'a_original' is affected!
print(f"Shallow copied list 'd' after modification: {d}\n")
````

````python
# Demonstrating Unpacking Multiple Lists
a = [1, 2]
c = [3, 4]
b = [*a, *c]
print(f"List a: {a}")
print(f"List c: {c}")
print(f"New list b created by unpacking a and c: {b}")
````

````python
import copy

# Original list with nested dictionaries
original_list = [{'id': 1, 'data': {'value': 10}}, {'id': 2, 'data': {'value': 20}}]
print(f"Original list: {original_list}\n")

# --- Shallow Copy with Nested Dictionaries ---
print("--- Demonstrating Shallow Copy with Nested Dictionaries ---")
shallow_copy = original_list.copy() # Or list(original_list) or original_list[:]
print(f"Shallow copied list (initially): {shallow_copy}")

# Modify a nested dictionary in the shallow copy
shallow_copy[0]['data']['value'] = 999
print(f"Modified 'shallow_copy[0][\"data\"][`value`]' to 999.")

print(f"Original list after shallow copy modification: {original_list}") # Original IS AFFECTED!
print(f"Shallow copied list after modification: {shallow_copy}\n")

# --- Deep Copy with Nested Dictionaries ---
print("--- Demonstrating Deep Copy with Nested Dictionaries ---")
deep_copy = copy.deepcopy(original_list)
print(f"Deep copied list (initially): {deep_copy}")

# Modify a nested dictionary in the deep copy
deep_copy[0]['data']['value'] = 888
print(f"Modified 'deep_copy[0][\"data\"][`value`]' to 888.")

print(f"Original list after deep copy modification: {original_list}") # Original is NOT AFFECTED!
print(f"Deep copied list after modification: {deep_copy}")
````
