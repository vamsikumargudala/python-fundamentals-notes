# **Tuples**

A tuple is a collection of different data types which is ordered and unchangeable (immutable). Tuples are written with round brackets, (). Once a tuple is created, we cannot change its values. We cannot use add, insert, remove methods in a tuple because it is not modifiable (mutable). Unlike list, tuple has few methods. Methods related to tuples:

*   `tuple()`: to create an empty tuple
*   `count()`: to count the number of a specified item in a tuple
*   `index()`: to find the index of a specified item in a tuple
*   `operator`: to join two or more tuples and to create a new tuple

## Accessing Tuple Items
### Positive Indexing
Similar to the list data type we use positive or negative indexing to access tuple items. Accessing tuple items


```python

# Syntax
tpl = ('item1', 'item2', 'item3')
first_item = tpl[0]
second_item = tpl[1]
print(f"First item: {first_item}, Second item: {second_item}")

````python
fruits = ('banana', 'orange', 'mango', 'lemon')
first_fruit = fruits[0]
second_fruit = fruits[1]
last_index = len(fruits) - 1
last_fruit = fruits[last_index]
print(f"First fruit: {first_fruit}, Second fruit: {second_fruit}, Last fruit: {last_fruit}")
````

### Negative indexing
Negative indexing means beginning from the end, -1 refers to the last item, -2 refers to the second last and the negative of the list/tuple length refers to the first item. Tuple Negative indexing


```python
# Syntax
tpl = ('item1', 'item2', 'item3','item4')
first_item = tpl[-4]
second_item = tpl[-3]
print(f"First item (negative index): {first_item}, Second item (negative index): {second_item}")


````python

fruits = ('banana', 'orange', 'mango', 'lemon')
first_fruit = fruits[-4]
second_fruit = fruits[-3]
last_fruit = fruits[-1]
print(f"First fruit (negative index): {first_fruit}, Second fruit (negative index): {second_fruit}, Last fruit (negative index): {last_fruit}")
````

## Slicing tuples
We can slice out a sub-tuple by specifying a range of indexes where to start and where to end in the tuple, the return value will be a new tuple with the specified items.

## Range of Positive Indexes

```python
# Syntax
tpl = ('item1', 'item2', 'item3','item4')
all_items = tpl[0:4]         # all items
all_items = tpl[0:]         # all items
middle_two_items = tpl[1:3]  # does not include item at index 3
```

````python
fruits = ('banana', 'orange', 'mango', 'lemon')
all_fruits = fruits[0:4]    # all items
print(f"All fruits (0:4): {all_fruits}")

all_fruits = fruits[0:]      # all items
print(f"All fruits (0:): {all_fruits}")

orange_mango = fruits[1:3]  # doesn't include item at index 3
print(f"Orange and Mango: {orange_mango}")

orange_to_the_rest = fruits[1:]
print(f"Orange to the rest: {orange_to_the_rest}")
````

## Range of Negative Indexes

```python
# Syntax
tpl = ('item1', 'item2', 'item3','item4')
all_items = tpl[-4:]         # all items
middle_two_items = tpl[-3:-1]  # does not include item at index 3 (-1)
```

````python
fruits = ('banana', 'orange', 'mango', 'lemon')
all_fruits = fruits[-4:]    # all items
print(f"All fruits (negative index -4:): {all_fruits}")

orange_mango = fruits[-3:-1]  # doesn't include item at index -1
print(f"Orange and Mango (negative index): {orange_mango}")

orange_to_the_rest = fruits[-3:]
print(f"Orange to the rest (negative index): {orange_to_the_rest}")
````

## Changing Tuples to Lists
We can change tuples to lists and lists to tuples. Tuple is immutable, so if we want to modify a tuple, we should change it to a list.

```python
# Syntax
tpl = ('item1', 'item2', 'item3','item4')
lst = list(tpl)
```

````python
fruits = ('banana', 'orange', 'mango', 'lemon')
print(f"Original tuple: {fruits}")
fruits_list = list(fruits)
fruits_list[0] = 'apple'
print(f"Modified list: {fruits_list}")     # ['apple', 'orange', 'mango', 'lemon']
fruits = tuple(fruits_list)
print(f"Converted back to tuple: {fruits}")     # ('apple', 'orange', 'mango', 'lemon')
````

## Checking an Item in a Tuple
We can check if an item exists or not in a tuple using `in`, which returns a boolean.

```python
# Syntax
tpl = ('item1', 'item2', 'item3','item4')
'item2' in tpl # True
```

````python
fruits = ('banana', 'orange', 'mango', 'lemon')
print(f"Is 'orange' in fruits? {'orange' in fruits}") # True
print(f"Is 'apple' in fruits? {'apple' in fruits}") # False
# The following line will cause a TypeError, demonstrating immutability:
# fruits[0] = 'apple' # TypeError: 'tuple' object does not support item assignment
````

## Deleting Tuples

While individual items within a tuple cannot be deleted (due to tuples being immutable), an entire tuple can be removed from memory using the `del` keyword. Once deleted, any attempt to access the tuple will result in a `NameError`.

**Reasoning**:
Now, I need to add a code cell to demonstrate the deletion of a tuple using `del` and show that accessing it afterwards will raise a `NameError`, as specified in the instructions.



````python
my_tuple = ('apple', 'banana', 'cherry')
print(f"Original tuple: {my_tuple}")

del my_tuple
print("Tuple deleted.")

# Attempting to access the tuple after deletion will raise a NameError
try:
    print(my_tuple)
except NameError as e:
    print(f"Error: {e}. The tuple 'my_tuple' no longer exists.")
````

## Exercise: Slice Middle Items

### Subtask:
Slice out the middle item or items from either 'food_stuff_tp' or 'food_stuff_lt'.


I will slice out the middle item or items from the `food_stuff_lt`.
First, I need to determine its length to correctly identify the middle index or indices. If the length is even, there will be two middle items; if odd, there will be one. Then, I will use Python's slicing capabilities to extract these item(s) and store them in a new variable, `middle_items`. Finally, I will print `middle_items` to display the result, fulfilling all parts of the subtask.

**Reasoning**:
Now that the explanation for the subtask is in place, I will add a code cell to implement the logic for slicing out the middle item(s) from `food_stuff_lt`.



````python
list_length = len(food_stuff_lt)

if list_length % 2 == 0: # Even number of items
    # Two middle items
    middle_index_1 = list_length // 2 - 1
    middle_index_2 = list_length // 2
    middle_items = food_stuff_lt[middle_index_1 : middle_index_2 + 1]
else: # Odd number of items
    # One middle item
    middle_index = list_length // 2
    middle_items = [food_stuff_lt[middle_index]] # Return as a list for consistency

print(f"Original list: {food_stuff_lt}")
print(f"Length of list: {list_length}")
print(f"Middle item(s): {middle_items}")
````
