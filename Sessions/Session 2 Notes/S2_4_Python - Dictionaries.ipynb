Dictionaries are the structures which holds the data in key - value pair format. For example, key can be student id and value can be student object corresponding to that student id. It allows faster access to the object as compared to the elements stored as part of list.

Dictionaries are more general form of the lists. They are more readable and one does not to keep the index in mind while accessing it. For example, a list of months is available, in order to access the October month, one need to keep its index in mind. Also if days of those months also needs to be referred, then another list needs to be maintained.

## Accessing Dictionary Items
We can access Dictionary items by referring to its key name.

### Syntax

````python
dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
print(dct['key1']) # value1
print(dct['key4']) # value4
````

### Example:

````python
person = {
    'first_name':'Asabeneh',
    'last_name':'Yetayeh',
    'age':250,
    'country':'Finland',
    'is_marred':True,
    'skills':['JavaScript', 'React', 'Node', 'MongoDB', 'Python'],
    'address':{
        'street':'Space street',
        'zipcode':'02210'
    }
    }
print(person['first_name']) # Asabeneh
print(person['country'])    # Finland
print(person['skills'])     # ['JavaScript', 'React', 'Node', 'MongoDB', 'Python']
print(person['skills'][0])  # JavaScript
print(person['address']['street']) # Space street
# print(person['city'])       # This would raise a KeyError as the key is not present
````

Accessing an item by key name raises an error if the key is not present, as shown in the commented out line `print(person['city'])`.

### Accessing Dictionary Items with `get()`
The `get()` method provides a safe way to access dictionary items. If the key is not found, it returns `None` instead of raising a `KeyError`.

````python
person = {
    'first_name':'Asabeneh',
    'last_name':'Yetayeh',
    'age':250,
    'country':'Finland',
    'is_marred':True,
    'skills':['JavaScript', 'React', 'Node', 'MongoDB', 'Python'],
    'address':{
        'street':'Space street',
        'zipcode':'02210'
    }
    }
print(person.get('first_name')) # Asabeneh
print(person.get('country'))    # Finland
print(person.get('skills')) # ['JavaScript', 'React', 'Node', 'MongoDB', 'Python']
print(person.get('city'))   # None
````

## Checking Keys in a Dictionary
We use the `in` operator to check if a key exists in a dictionary.

### Syntax

````python
dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
print('key2' in dct) # True
print('key5' in dct) # False
````

## Removing Key and Value Pairs from a Dictionary

- `pop(key)`: removes the item with the specified key name.
- `popitem()`: removes the last inserted item (in Python 3.7+; for older versions, it removes an arbitrary item).
- `del`: removes an item with a specified key name or deletes the entire dictionary.

### Syntax

````python
dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
dct.pop('key1') # removes key1 item
print(dct)
dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
dct.popitem() # removes the last item
print(dct)
dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
del dct['key2'] # removes key2 item
print(dct)
````

### Example:

````python
person = {
    'first_name':'Asabeneh',
    'last_name':'Yetayeh',
    'age':250,
    'country':'Finland',
    'is_marred':True,
    'skills':['JavaScript', 'React', 'Node', 'MongoDB', 'Python'],
    'address':{
        'street':'Space street',
        'zipcode':'02210'
    }
    }
person.pop('first_name')        # Removes the first_name item
print(person)
person.popitem()                # Removes the address item
print(person)
del person['is_marred']        # Removes the is_marred item
print(person)
````

## Changing Dictionary to a List of Items
The `items()` method changes a dictionary to a list of tuples, where each tuple is a key-value pair.

### Syntax

````python
dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
print(dct.items()) # dict_items([('key1', 'value1'), ('key2', 'value2'), ('key3', 'value3'), ('key4', 'value4')])
````

## Clearing a Dictionary
If we want to remove all items from a dictionary, we can use the `clear()` method.

### Syntax

````python
dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
dct.clear()
print(dct) # {}
````

## Deleting a Dictionary
If we no longer need the dictionary, we can delete it completely using the `del` keyword.

### Syntax

````python
dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
del dct
# print(dct) # This would raise a NameError as the dictionary no longer exists
````

## Copy a Dictionary
We can copy a dictionary using the `copy()` method. Using `copy()` ensures that changes to the new dictionary do not affect the original dictionary.

### Syntax

````python
dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
dct_copy = dct.copy()
print(dct_copy) # {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
````

## Getting Dictionary Keys as a List
The `keys()` method gives us all the keys of a dictionary as a list-like view object.

### Syntax

````python
dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
keys = dct.keys()
print(keys)     # dict_keys(['key1', 'key2', 'key3', 'key4'])
````

## Getting Dictionary Values as a List
The `values()` method gives us all the values of a dictionary as a list-like view object.

### Syntax

````python
dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
values = dct.values()
print(values)     # dict_values(['value1', 'value2', 'value3', 'value4'])
````

<b> Looping <B>

````python
#Dictionary with Roman numerals
letter_dict = { 'I':1, 'V':5, 'X':10, 'L':50, 'C':100, 'D':500, "M":1000}
letter_dict
````

````python
for key in letter_dict.keys():   # iterate over keys
    print("   " , key)
````

````python
for value in letter_dict.values(): # iterate over values
    print("   " , value)
````

````python
for key in letter_dict.keys():  #access keys and values
    print(key , "   " , letter_dict[key])
````

````python
#Q1 Solution

#Number of times asks the user to input the information
count = 5

#empty dictionary
products = {}

#Ask the user for product names and corresponding price, then add it to the dictionary
for i in range(count):
    product_name = input("Enter the product name\n")
    price = eval(input("Enter the product price\n"))
    products[product_name] = price


print("\n")
product_name = input("Enter the product name\n")
print("The corresponding price is ", products[product_name])
````

````python
#Q2 Solution
months = {'January':31, 'February':28, 'March':31, 'April':30, "May":31, 'June':30, 'July':31, "August":31, 'September':30,
          'October':31, 'November':30, 'December':31}

#(a) Ask the user to input a month name and use the dictionary to tell how many days are in that month.
month_name = input("Enter the month\n")
print(month_name, " has ", months[month_name], " days.")

#(b) Print out all the keys in alphabetical order.
print("\nThe keys in alphabetical order : \n")
for key in sorted(months.keys()):
    print(key)
````

````python
#Q3 Solution

#Number of times asks the user to input the information
count = 2

#empty dictionary
teams = {}

#Ask the user for team names and corresponding wins & losses, then add it to the dictionary
for i in range(count):
    team_name = input("Enter the team name\n")
    wins = int(input("Enter the wins\n"))
    losses = int(input("Enter the losses\n"))
    teams[team_name] = [wins, losses]


print("\n")
team_name = input("Enter the team name\n")
print("The winning percentage is ", teams[team_name][0] / (teams[team_name][0] + teams[team_name][1]))

print("\n")
winning_list = []
for value in teams.values():
    winning_list.append(value[0])

print('The winning list is ', winning_list)
````
