## 1. Pandas Series

A **Series** is a one-dimensional labeled array that can hold any data type.

Think of it as:
- A single column of data
- A dictionary with ordered keys (index)
- A NumPy array with labels

### 1.1 Creating Series from Different Sources

````python
# From a Python list
grades = [85, 90, 78, 92, 88]
series_list = pd.Series(grades)
print("Series from list:")
print(series_list)
print("\nData type:", series_list.dtype)
````

````python
# From a NumPy array with custom index
array = np.array([10, 20, 30, 40, 50])
index = ['a', 'b', 'c', 'd', 'e']
series_array = pd.Series(array, index=index)
print("Series from array with custom index:")
print(series_array)
````

````python
# From a dictionary (keys become index)
student_scores = {
    'Alice': 95,
    'Bob': 87,
    'Charlie': 92,
    'Diana': 88
}
series_dict = pd.Series(student_scores)
print("Series from dictionary:")
print(series_dict)
````

````python
# From a scalar (repeated for each index)
series_scalar = pd.Series(5.0, index=['x', 'y', 'z'])
print("Series from scalar:")
print(series_scalar)
````

### 1.2 Series Properties

````python
# Exploring Series properties
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'], name='my_series')

print("Values:", s.values)         # NumPy array of values
print("Index:", s.index)           # Index object
print("Data type:", s.dtype)       # Data type
print("Shape:", s.shape)           # Dimensions
print("Size:", s.size)             # Number of elements
print("Name:", s.name)             # Series name
````

### 1.3 Accessing Series Elements

````python
s = pd.Series([100, 200, 300, 400, 500], index=['a', 'b', 'c', 'd', 'e'])
print("Original Series:")
print(s)
print()

# Access by label
print("s['b'] =", s['b'])

# Access by position
print("s[1] =", s[1])

# Access multiple elements by labels
print("\ns[['a', 'c', 'e']] =")
print(s[['a', 'c', 'e']])

# Slicing by label (inclusive!)
print("\ns['b':'d'] =")
print(s['b':'d'])
````

## 2. Pandas DataFrame

A **DataFrame** is a 2-dimensional labeled data structure with columns of potentially different types.

Think of it as:
- A spreadsheet or SQL table
- A dictionary of Series objects
- A 2D NumPy array with row and column labels

### 2.1 Creating DataFrames

````python
# From a dictionary of lists
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'age': [25, 30, 35, 28],
    'city': ['NYC', 'LA', 'Chicago', 'Houston'],
    'salary': [50000, 60000, 75000, 55000]
}
df = pd.DataFrame(data)
print("DataFrame from dictionary:")
df
````

````python
# From a dictionary of Series
dict_series = {
    'name': pd.Series(['Alice', 'Bob', 'Charlie']),
    'score': pd.Series([85, 90, 78]),
    'passed': pd.Series([True, True, True])
}
df_series = pd.DataFrame(dict_series)
df_series
````

````python
# From a 2D NumPy array with custom index and columns
array_2d = np.random.randint(1, 100, size=(4, 3))
df_array = pd.DataFrame(
    array_2d,
    index=['row1', 'row2', 'row3', 'row4'],
    columns=['A', 'B', 'C']
)
print("DataFrame from NumPy array:")
df_array
````

````python
# From a list of dictionaries
records = [
    {'name': 'Alice', 'age': 25, 'dept': 'HR'},
    {'name': 'Bob', 'age': 30, 'dept': 'IT'},
    {'name': 'Charlie', 'age': 35, 'dept': 'Finance'}
]
df_records = pd.DataFrame(records)
df_records
````

### 2.2 DataFrame Properties

````python
# Create sample DataFrame
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'age': [25, 30, 35, 28, 22],
    'salary': [50000.0, 60000.0, 75000.0, 55000.0, 45000.0],
    'active': [True, True, False, True, True]
})

print("DataFrame:")
print(df)
print()

print("Shape:", df.shape)           # (rows, columns)
print("Size:", df.size)             # Total elements
print("Columns:", list(df.columns)) # Column names
print("Index:", list(df.index))     # Row labels
print("\nData types:")
print(df.dtypes)
````

### 2.3 Viewing Data: head(), tail(), info()

````python
# First n rows (default 5)
print("Head (first 3 rows):")
print(df.head(3))
````

````python
# Last n rows (default 5)
print("Tail (last 2 rows):")
print(df.tail(2))
````

````python
# DataFrame info - columns, types, memory usage
df.info()
````

## 3. Data Selection & Indexing

There are multiple ways to select data from a DataFrame:

| Method | Description |
|--------|-------------|
| `df['col']` | Select single column (returns Series) |
| `df[['col1', 'col2']]` | Select multiple columns (returns DataFrame) |
| `df.loc[]` | Label-based selection |
| `df.iloc[]` | Integer position-based selection |

### 3.1 Column Selection

````python
# Single column (returns Series)
print("Single column (df['name']):")
print(df['name'])
print("Type:", type(df['name']))
````

````python
# Alternative: dot notation (only works for valid Python identifiers)
print("Using dot notation (df.age):")
print(df.age)
````

````python
# Multiple columns (returns DataFrame)
print("Multiple columns:")
print(df[['name', 'salary']])
print("Type:", type(df[['name', 'salary']]))
````

### 3.2 Row Selection with .loc (Label-based)

````python
# Select single row by label
print("Row at index 0:")
print(df.loc[0])
````

````python
# Select range of rows (inclusive on both ends!)
print("Rows 1 to 3:")
print(df.loc[1:3])
````

````python
# Select specific rows and columns
print("Rows 0-2, columns 'name' and 'age':")
print(df.loc[0:2, ['name', 'age']])
````

````python
# Select single cell
print("Cell at row 1, column 'name':")
print(df.loc[1, 'name'])
````

### 3.3 Row Selection with .iloc (Position-based)

````python
# Select by position (like array indexing)
print("First row (position 0):")
print(df.iloc[0])
````

````python
# Slice rows (exclusive end!)
print("Rows 0 to 2 (positions):")
print(df.iloc[0:3])
````

````python
# Select specific rows and columns by position
print("Rows 0-1, columns 0-1:")
print(df.iloc[0:2, 0:2])
````

````python
# Select every other row
print("Every other row:")
print(df.iloc[::2])
````

### 3.4 Filtering with Boolean Conditions

````python
# Filter rows where age > 25
print("Employees older than 25:")
print(df[df['age'] > 25])
````

````python
# Multiple conditions with & (and) and | (or)
print("Age > 25 AND salary > 55000:")
print(df[(df['age'] > 25) & (df['salary'] > 55000)])
````

````python
# Negation with ~
print("Not active employees:")
print(df[~df['active']])
````

````python
# Using isin() for multiple values
print("Names in list ['Alice', 'Charlie']:")
print(df[df['name'].isin(['Alice', 'Charlie'])])
````

````python
# Using query() method (SQL-like syntax)
print("Using query():")
print(df.query('age > 25 and salary > 50000'))
````

---

## Key Differences: .loc vs .iloc

| Feature | `.loc` | `.iloc` |
|---------|--------|--------|
| Based on | Labels | Integer positions |
| End point | **Inclusive** | **Exclusive** |
| Column access | By name | By position |

```python
df.loc[0:2]   # Rows 0, 1, 2 (inclusive)
df.iloc[0:2]  # Rows 0, 1 (exclusive of 2)
```

### Solution 1: Create a Series

````python
# Solution 1
temperatures = pd.Series(
    [72, 75, 68, 70, 74],
    index=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    name='temperatures'
)
print("Temperatures Series:")
print(temperatures)
print()

# Temperature on Wednesday
print("Wednesday temperature:", temperatures['Wednesday'])

# Tuesday to Thursday (inclusive with loc)
print("\nTuesday to Thursday:")
print(temperatures['Tuesday':'Thursday'])
````

### Solution 2: Create a DataFrame

````python
# Solution 2
students = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'quiz1': [8, 7, 9, 6, 8],
    'quiz2': [9, 8, 7, 8, 9],
    'midsem': [85, 78, 92, 75, 88],
    'compre': [90, 82, 88, 80, 95]
})

print("Students DataFrame:")
print(students)
print()

# First 3 rows
print("First 3 rows:")
print(students.head(3))
print()

# Column names and data types
print("Columns:", list(students.columns))
print("\nData types:")
print(students.dtypes)
````

### Solution 3: Selection Practice

````python
# Solution 3

# 1. Select name and compre columns
print("1. Name and Compre columns:")
print(students[['name', 'compre']])
print()

# 2. Rows where midsem > 80
print("2. Students with midsem > 80:")
print(students[students['midsem'] > 80])
print()

# 3. quiz1 >= 8 AND compre >= 90
print("3. Students with quiz1 >= 8 AND compre >= 90:")
print(students[(students['quiz1'] >= 8) & (students['compre'] >= 90)])
print()

# 4. Last 2 rows using iloc
print("4. Last 2 rows:")
print(students.iloc[-2:])
````

### Solution 4: DataFrame with DateTime Index

````python
# Solution 4
np.random.seed(42)  # For reproducibility

# Create datetime index
date_index = pd.date_range('2024-01-01', periods=7, freq='D')

# Create DataFrame
weather_df = pd.DataFrame({
    'temperature': np.random.randint(60, 80, size=7),
    'humidity': np.random.randint(40, 90, size=7),
    'wind_speed': np.random.uniform(5, 20, size=7).round(1)
}, index=date_index)

print("Weather DataFrame with DateTime Index:")
print(weather_df)
print()
print("Index type:", type(weather_df.index))
````
