## While Loop

Repeats while condition is True. Must update loop variable to avoid infinite loop.

```python
while condition:
    # code
    # update loop variable!
```

````python
count = 0
while count < 5:
    print(count)
    count += 1  # Important: update the variable!
````

````python
# while-else: else runs when condition becomes False
count = 0
while count < 3:
    print(count)
    count += 1
else:
    print('Loop finished normally')
````

## For Loop

Iterate over sequences (list, tuple, string, dict, set, range).

```python
for item in sequence:
    # code using item
```

````python
# Iterate over list
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)
````

````python
# Iterate over string
for char in 'Python':
    print(char)
````

````python
# Iterate over dictionary
person = {'name': 'Alice', 'age': 25}

# Keys only (default)
for key in person:
    print(key)

print('---')

# Key-value pairs
for key, value in person.items():
    print(f'{key}: {value}')
````

## range() Function

Generate sequences of numbers.

```python
range(stop)              # 0 to stop-1
range(start, stop)       # start to stop-1
range(start, stop, step) # with custom step
```

````python
# range(5) -> 0, 1, 2, 3, 4
for i in range(5):
    print(i)
````

````python
# range(1, 6) -> 1, 2, 3, 4, 5
for i in range(1, 6):
    print(i)
````

````python
# range(0, 10, 2) -> 0, 2, 4, 6, 8 (step of 2)
for i in range(0, 10, 2):
    print(i)
````

````python
# Countdown: range(5, 0, -1) -> 5, 4, 3, 2, 1
for i in range(5, 0, -1):
    print(i)
````

## enumerate() - Get Index and Value

When you need both the index and the value.

````python
colors = ['red', 'green', 'blue']

for index, color in enumerate(colors):
    print(f'{index}: {color}')
````

````python
# Start from custom index
for index, color in enumerate(colors, start=1):
    print(f'{index}: {color}')
````

## break and continue

- `break` - exit the loop immediately
- `continue` - skip to next iteration

````python
# break: stop when found
for i in range(10):
    if i == 5:
        print('Found 5, stopping!')
        break
    print(i)
````

````python
# continue: skip even numbers
for i in range(6):
    if i % 2 == 0:
        continue  # skip to next iteration
    print(i)  # only prints odd: 1, 3, 5
````

## Infinite Loop Warning

Always update your loop variable in while loops!

```python
# BAD - infinite loop!
i = 0
while i < 5:
    print(i)  # i never changes, loops forever

# GOOD
i = 0
while i < 5:
    print(i)
    i += 1    # update the variable
```

## Common Patterns

````python
# Sum all items
numbers = [1, 2, 3, 4, 5]
total = 0
for num in numbers:
    total += num
print(f'Sum: {total}')
````

````python
# Find maximum
numbers = [3, 1, 4, 1, 5, 9, 2]
max_val = numbers[0]
for num in numbers:
    if num > max_val:
        max_val = num
print(f'Max: {max_val}')
````

````python
# Build a list
squares = []
for i in range(1, 6):
    squares.append(i ** 2)
print(squares)
````

## Nested Loops

````python
# Multiplication table
for i in range(1, 4):
    for j in range(1, 4):
        print(f'{i} x {j} = {i*j}')
    print('---')
````

````python
# Print triangle pattern
rows = 5
for i in range(1, rows + 1):
    print('*' * i)
````

## Quick Reference

| Construct | Use Case |
|-----------|----------|
| `for x in list` | Iterate over items |
| `for i in range(n)` | Repeat n times |
| `for i, x in enumerate(list)` | Index + value |
| `while condition` | Repeat until False |
| `break` | Exit loop early |
| `continue` | Skip to next iteration |
| `range(start, stop, step)` | Generate number sequence |
