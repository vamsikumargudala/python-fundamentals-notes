## Input is Always a String

Even if the user types a number, `input()` returns a string.

````python
age = input('Enter your age: ')
print('Age:', age)
print('Type:', type(age))  # <class 'str'>
````

## Type Conversion

Convert input to the required type:

| Function | Converts to |
|----------|------------|
| `int()` | Integer |
| `float()` | Decimal |
| `str()` | String |

````python
# Convert to integer
age = input('Enter your age: ')
age = int(age)
print('Type:', type(age))  # <class 'int'>
````

````python
# Convert to float
salary = input('Enter salary: ')
salary = float(salary)
print('Type:', type(salary))  # <class 'float'>
````

````python
# One-liner conversion
age = int(input('Enter your age: '))
salary = float(input('Enter salary: '))
````

## Using eval()

`eval()` evaluates a string as Python code. It auto-detects the type.

````python
# Automatically becomes int
age = eval(input('Enter age: '))  # User types: 25
print(type(age))  # <class 'int'>
````

````python
# Automatically becomes float
salary = eval(input('Enter salary: '))  # User types: 50000.50
print(type(salary))  # <class 'float'>
````

````python
# Can even evaluate expressions
result = eval(input('Enter expression: '))  # User types: 10 + 20 * 3
print(result)  # 70
````

**Warning:** `eval()` is risky with untrusted input - it executes any Python code!

````python
# Q1: Ask for a number, print its square and cube

num = int(input('Enter a number: '))
print('Square:', num ** 2)
print('Cube:', num ** 3)
````

````python
# Q2: Input 3 numbers, calculate total and average

n1 = float(input('Enter first number: '))
n2 = float(input('Enter second number: '))
n3 = float(input('Enter third number: '))

total = n1 + n2 + n3
avg = total / 3

print('Numbers:', n1, n2, n3)
print('Total:', total)
print('Average:', avg)
````

````python
# Q3: Salary estimator
# HRA = 10% of basic, Conveyance = 3500

basic = float(input('Enter basic salary: '))
hra = basic * 0.10
conveyance = 3500
total_salary = basic + hra + conveyance

print('Basic:', basic)
print('HRA (10%):', hra)
print('Conveyance:', conveyance)
print('Total Salary:', total_salary)
````
